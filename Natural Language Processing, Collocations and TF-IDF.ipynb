```python
# modify these two functions
import nltk
import csv
from nltk.corpus import stopwords
import string
import pandas as pd
import numpy as np
from sklearn.preprocessing import normalize
from nltk.stem.porter import PorterStemmer
from scipy.spatial import distance
from collections import Counter
import nltk, re, string
from sklearn.preprocessing import normalize
from nltk.corpus import stopwords
# numpy is the package for matrix cacluation
import numpy as np  
import pandas as pd
from nltk.stem.porter import PorterStemmer

#  get tokens of each document as list
def get_doc_tokens(doc,normalize):
    stop_words = nltk.corpus.stopwords.words('english')
    tokens=[token.strip() \
            for token in nltk.word_tokenize(doc.lower()) \
            if token.strip() not in stop_words and\
               token.strip() not in string.punctuation\
               and token.strip().isalpha()]
    porter_stemmer = PorterStemmer()
    for xx in tokens:
        if normalize=='stem':
            print(porter_stemmer.stem(xx))
    
    
    token_count={token:tokens.count(token) for token in set(tokens)}
    return token_count

def tfidf(docs,normalize):
    #  processing all documents to get list of token list
    docs_tokens={idx:get_doc_tokens(doc,normalize) for idx,doc in enumerate(docs)}

    #  getting document-term matrix
    dtm=pd.DataFrame.from_dict(docs_tokens, orient="index" )
    dtm=dtm.fillna(0)
      
    # getting normalized term frequency (tf) matrix        
    tf=dtm.values
    doc_len=tf.sum(axis=1)
    tf=np.divide(tf.T, doc_len).T
    
    #  get idf
    df=np.where(tf>0,1,0)
    smoothed_idf=np.log(np.divide(len(docs)+1, np.sum(df, axis=0)+1))+1    
    smoothed_tf_idf=tf*smoothed_idf
    
    return smoothed_tf_idf
```

```python
def top_collocation(tokens, K):
    Result=[]
    
    
    tagged_tokens= nltk.pos_tag(tokens)

    bigrams=list(nltk.bigrams(tagged_tokens))
    result=[ (x[0],y[0]) for (x,y) in bigrams \
             if x[1].startswith('JJ') and y[1].startswith('NN') or x[1].startswith('NN') and y[1].startswith('NN') ]
    Result = nltk.FreqDist(result).most_common(K)


    return Result
```
